```python
plt.figure(figsize=(9,5))
sns.boxplot(x = 'member_casual', y='bike_rides_daily', data=df, palette={'member': 'skyblue', 'casual': 'lightcoral'})

# Customize the axis labels
plt.title('Box Plot of Ride Duration by Membership Type', fontsize=18)
plt.xlabel('Membership Type', fontsize=14)
plt.ylabel('Daily Bike Rides', fontsize=14)
plt.savefig('box_plot_member_trip_duration.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
plt.show()
```

This box plot compares the daily bike rides between two membership types: "member" and "casual." The y-axis represents the number of daily bike rides, ranging from 0 to 1400. The box plot for members shows a median daily bike ride count of around 1000, with the interquartile range (IQR) spanning from approximately 750 to 1200. For casual users, the median is slightly higher, around 1100, with the IQR ranging from about 900 to 1200. The plot also indicates that casual users have more variability in their daily bike rides, as evidenced by the presence of several outliers below the lower whisker. It's interesting to see that casual rider's have a higher number of rides daily.

#### Create a FacetGrid and write 2 to 3 sentences analyzing what insight its output brings.

```python
df.columns
```

```python
# Convert 'started_at' and 'ended_at' to datetime
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# Calculate the trip duration in minutes
df['trip_duration'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60

# Verify that the 'tripduration' column has been created successfully
df[['started_at', 'ended_at', 'trip_duration']]
```

```python
#Remove outliers in 'trip_duration'
#Calculate the IQR
Q1 = df['trip_duration'].quantile(0.25)
Q3 = df['trip_duration'].quantile(0.75)
IQR = Q3 - Q1

# Filter the DataFrame to remove outliers
df_filtered = df[(df['trip_duration'] >= (Q1 - 1.5 * IQR)) & (df['trip_duration'] <= (Q3 + 1.5 * IQR))]
```

```python
df_filtered
```

```python
# Create a FacetGrid plot to visualize the distribution of trip duration by membership type and rideable type
grid= sns.FacetGrid(df_filtered, col='member_casual', row='rideable_type', margin_titles=True, height=3, aspect=1.5)
grid.map(sns.histplot, 'trip_duration', bins=10, kde=True, color='skyblue')

# Add titles, labels, and adjust layout
grid.set_axis_labels('Trip Duration (minutes)', 'Frequency')
grid.set_titles(col_template='{col_name}', row_template='{row_name}')
grid.fig.suptitle('Distribution of Trip Duration by Membership Type and Rideable Type', fontsize=16)
grid.fig.subplots_adjust(top=0.9)

plt.show()
```

The FacetGrid plot reveals distinct patterns in trip durations based on membership type and rideable type. For members using both classic and electric bikes, shorter trip durations are more common, indicating that members may prefer quick commutes. On the other hand, casual users show a more varied distribution of trip durations, especially with electric bikes, which suggests that they might be using the service for a broader range of activities, including leisure and exploration.
